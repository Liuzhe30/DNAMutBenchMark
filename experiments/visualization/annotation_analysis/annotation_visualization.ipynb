#### annotation visualization

```python
import umap
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

variant_id = ['chr17_8106013_A_C_b38','chr19_1364856_T_C_b38','chr2_113185893_A_G_b38']
```

#### GPN

```python
embedding_path = '../../../datasets/benchmark_datasets/'
annotation_path = '../../../datasets/benchmark_datasets/annotation_501bp.pkl'
annotation_table = pd.read_pickle(annotation_path)

embedding = 'gpn'
embedding_data = pd.read_pickle(embedding_path + embedding + '.dataset')

variant = 'chr17_8106013_A_C_b38'
embedding_before = embedding_data[embedding_data['variant_id']==variant]['embedding_before'].values[0][0]
embedding_after = embedding_data[embedding_data['variant_id']==variant]['embedding_after'].values[0][0]
annotation = np.array(np.array(annotation_table[annotation_table['variant_id']==variant]['annotation'])[0])

embedding_mean = embedding_before.mean(axis=1)
df_trend = pd.DataFrame({
'Position': np.arange(1, 502),
'Embedding Mean': embedding_mean,
'Annotation': annotation
})
dim_names = [f'Dim_{i+1}' for i in range(embedding_before.shape[1])]
embedding_df = pd.DataFrame(embedding_before, columns=dim_names)
embedding_df['Annotation'] = annotation

plt.figure(figsize=(4, 3))

custom_palette = {
'distal_enhancer': '#8FBC8F',      
'null': '#6A5ACD'   
}

for dim in dim_names:
    sns.kdeplot(data=embedding_df, x=dim, hue='Annotation', fill=True, common_norm=False, alpha=0.1, palette=custom_palette)
plt.title('Kernel Density Plot of GPN\nBefore Mutation,'+variant, fontsize=12)
plt.xlabel('Embedding Value per Dimension', fontsize=12)
plt.ylabel('Probability Density (KDE)', fontsize=12)
plt.legend(title='Region Type', loc='upper right', labels=['Enhancer', 'Non-Enhancer'])
plt.grid(alpha=0.5, linestyle='--')
plt.tight_layout()
plt.savefig('images/kdeplot-' + embedding + '-before-mutation-' + variant + '.png',dpi=300, bbox_inches = 'tight')
plt.show()

# after mutation
embedding_mean = embedding_after.mean(axis=1)
df_trend = pd.DataFrame({
'Position': np.arange(1, 502),
'Embedding Mean': embedding_mean,
'Annotation': annotation
})
dim_names = [f'Dim_{i+1}' for i in range(embedding_after.shape[1])]
embedding_df = pd.DataFrame(embedding_after, columns=dim_names)
embedding_df['Annotation'] = annotation

plt.figure(figsize=(4, 3))

custom_palette = {
'distal_enhancer': '#8FBC8F',      
'null': '#6A5ACD'   
}

for dim in dim_names:
    sns.kdeplot(data=embedding_df, x=dim, hue='Annotation', fill=True, common_norm=False, alpha=0.1, palette=custom_palette)
plt.title('Kernel Density Plot of GPN\nAfter Mutation,'+variant, fontsize=12)
plt.xlabel('Embedding Value per Dimension', fontsize=12)
plt.ylabel('Probability Density (KDE)', fontsize=12)
plt.legend(title='Region Type', loc='upper right', labels=['Enhancer', 'Non-Enhancer'])
plt.grid(alpha=0.5, linestyle='--')
plt.tight_layout()
plt.savefig('images/kdeplot-' + embedding + '-after-mutation-' + variant + '.png',dpi=300, bbox_inches = 'tight')
plt.show()
```
