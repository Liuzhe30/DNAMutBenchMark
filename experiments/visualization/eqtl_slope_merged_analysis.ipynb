```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

data = pd.read_csv('eqtl_slope_merged_result.csv')

plt.figure(figsize=(10, 3))

metrics = ['rmse', 'r2', 'pcc']

palette = sns.color_palette("twilight", as_cmap=False)

p_values = {}

for i, metric in enumerate(metrics):
    plt.subplot(1, 3, i + 1)
    
    shuffled_yes = data[data['shuffled'] == 'yes'][metric]
    shuffled_no = data[data['shuffled'] == 'no'][metric]

    t_stat, p_val = ttest_ind(shuffled_yes, shuffled_no)
    p_values[metric] = p_val

    n_yes = len(shuffled_yes)
    n_no = len(shuffled_no)

    ax = sns.boxplot(x='shuffled', y=metric, data=data, palette=palette, linewidth=1.2)

    for i, box in enumerate(ax.artists):
        
        median = box.get_ydata()[0]  
        upper = box.get_ydata()[2]  
        lower = box.get_ydata()[1]  
        
        ax.errorbar(i, median, yerr=[[median - lower], [upper - median]], fmt='o', color='black', capsize=4, label='Error bars')
    
    plt.title('average ' + f'{metric.capitalize()}\n'
              #f'({n_yes} samples in "yes", {n_no} samples in "no")\n'
              f'T-test, p<0.0001', fontsize=14)
    plt.xlabel('Shuffled?', fontsize=12)
    #plt.ylabel(f'{metric.capitalize()}', fontsize=12)

    plt.grid(alpha=0.5, linestyle='--')

plt.tight_layout()
plt.savefig('eqtl-slope-images/shuffled-all.png',dpi=300, bbox_inches = 'tight')
plt.show()

plt.figure(figsize=(11, 3.5))
for i, metric in enumerate(metrics):
    plt.subplot(1, 3, i + 1)
    
    shuffled_yes = data[data['shuffled'] == 'yes'][metric]
    shuffled_no = data[data['shuffled'] == 'no'][metric]

    t_stat, p_val = ttest_ind(shuffled_yes, shuffled_no)
    p_values[metric] = p_val

    n_yes = len(shuffled_yes)
    n_no = len(shuffled_no)

    ax = sns.boxplot(x='shuffled', y=metric, data=data, palette=palette, linewidth=1.2)

    for i, box in enumerate(ax.artists):
        
        median = box.get_ydata()[0]  
        upper = box.get_ydata()[2]  
        lower = box.get_ydata()[1]  
        
        ax.errorbar(i, median, yerr=[[median - lower], [upper - median]], fmt='o', color='black', capsize=4, label='Error bars')
    
    plt.title('average ' + f'{metric.capitalize()}\n'
              f'({n_yes} samples in "yes",\n {n_no} samples in "no")\n'
              f'T-test, p<0.0001', fontsize=14)
    plt.xlabel('Shuffled?', fontsize=12)
    plt.ylabel(f'{metric.capitalize()}', fontsize=12)

    plt.grid(alpha=0.5, linestyle='--')
    plt.tight_layout()
    plt.savefig('eqtl-slope-images/shuffled-all2.png',dpi=300, bbox_inches = 'tight')

print("P-values for each metric:")
for metric, p_val in p_values.items():
    print(f"{metric.capitalize()}: {p_val:.3f}")
```
